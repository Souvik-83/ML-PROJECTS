```python
#Encoding
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':0}},inplace=True)
```

```python
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
df
```

```python
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
df
```

```python
df.info()
```

```python
#splitting
X=df.drop(['Car_Name','Selling_Price'],axis=1)
Y=df['Selling_Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_test
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics # for accuracy and error calculation
```

```python
# load the model
lr= LinearRegression()
```

```python
lr
```

```python
lr.fit(X_train, Y_train)/
```

```python
train_pred=lr.predict(X_train)
test_pred=lr.predict(X_test)
```

```python
test_pred
```

```python
#R squared data
error_train=metrics.r2_score(Y_train,train_pred)
error_train
```

```python
error_test=metrics.r2_score(Y_test,test_pred)
error_test
```

```python
plt.scatter(train_pred, Y_train)
plt.xlabel("Actual price")
plt.ylabel("Predicted Price")
plt.show()
```
